```python
import glob
import json
def read_json(path):
    """폴더의 모든 JSON 파일을 읽어 하나의 리스트로 반환"""
    all_data = []
    
    json_files = glob.glob('./data/'+path+'/*.json')
    
    for file_path in json_files:
        try:
            # utf-8-sig를 사용하면 BOM을 자동으로 제거
            with open(file_path, 'r', encoding='utf-8-sig') as f:
                data = json.load(f)
                all_data.append(data['content'])
        except json.JSONDecodeError as e:
            print(f"❌ {file_path} - JSON 형식 오류: {e}")
        except FileNotFoundError:
            print(f"❌ {file_path} - 파일 없음")
    
    return all_data
```

```python
etc_data = read_json('KR_etc')
print(f"\n읽은 파일 수: {len(etc_data)}")
```

```python
textbook_data = read_json('KR_textbook')
print(f"\n읽은 파일 수: {len(textbook_data)}")
```
